```python
#Task5
#Generate the following metrics from the dataset
#create list for generated metrics
import csv
with open("myDatafiles\myfoodhygienedata.csv","r") as csvfile2:
    readcsv2= csv.DictReader(csvfile2, delimiter=",")
    ratings=[0,0,0,0,0,0]
   
    for row in readcsv2:
        if row["rating"] in (" " or None):
            ratings[0]+=1
        if row["rating"]=="1":
            ratings[1]+=1
        if row["rating"]=="2":
            ratings[2]+=1
        if row["rating"]=="3":
            ratings[3]+=1
        if row["rating"]=="4":
            ratings[4]+=1
        if row["rating"]=="5":
            ratings[5]+=1
     
    print("Metric of ratings, where a score of 0 is an empty/no rating")
    for rate in range(0,6):
        print("amount scoring {0}: {1}".format(rate,ratings[rate]))
```
