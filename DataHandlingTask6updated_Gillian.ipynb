```python
#Task6-updated with headers
# need to work out how to remove the 0:00:00

#Write a function which will: Read in the establishment names and inspection dates listed in the Food Hygiene Dataset
#For each record, calculate the value “DaysSinceInspection”. This value contains the number of days since the date specified in the “inspectiondate” field.
#Note: You will need to use the string “split” method on each date from the CSV file, 
#and then create a new Date object from these string components. 
#You will then need to subtract this new Date object from today’s date to find the difference.  
#You will also need to deal appropriately with records that have no date value (use an if statement)
#Create a new CSV file which contains a collection of establishment names, inspection dates, and days since inspection
import csv
import datetime

#date1 = datetime.datetime.today() - this will give me time too
todayDate=datetime.date.today()

with open("myDatafiles\myfoodhygienedata.csv","r") as foodfile:
    hygieneData= csv.DictReader(foodfile, delimiter=",")
    #create lists to read data too
    names=[]
    ispDates=[]
    dateDiffList=[]
    for row in hygieneData:
        names.append(row["establishmentname"])
        ispDates.append(row["inspectiondate"])
        
        if row["inspectiondate"]=="":
            dateDiff=0
        else:
            dateIns=datetime.datetime.strptime(row["inspectiondate"],'%d/%m/%Y').date()  #.date() just gives date(not time)
            dateDiff= (todayDate-dateIns)
            
        dateDiffList.append(dateDiff)
    
    #dict comprehsion to exclude where inspection date does not exist
    dict_results={name:datediff for name,datediff in zip(names,dateDiffList) if dateDiff!=0}
    #print(dict_results)

        
    #results file
    with open("myDatafiles\inspResult3.csv",'w', newline="") as file:
        header=["establishmentname", "datesinceinspection"]
        writer=csv.writer(file,delimiter=',')
        writer.writerow(header)
        writer.writerows(dict_results.items())
             
                         
         
```
